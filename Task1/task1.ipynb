**IMPORT PANDAS**

```python
import pandas as pd
```

**DATA READING**

```python
data_read = pd.read_csv('Shopping Mall Customer Segmentation Data .csv')
```

```python
data_read
```

**CHECKING NULL VALUES**


```python
data_read.info()
```

**REMOVING DUPLICATES**

```python
data_read = data_read.drop_duplicates()
data_read
```

**CLEANING AGE COLUMN**

```python
data_read['Age'] = data_read['Age'].astype(str).str.replace('[a-zA-Z]','')
```

**REPLACING NULL VALUES AS UNKNOWN**

```python
data_read = data_read.fillna('Unknown')
```

**ARRANGING AGE IN ASSENDING ORDER**

```python
data_read = data_read.sort_values(by='Age', ascending=True)
data_read
```

**CREATING AGE GROUP**

```python
bins = [0, 25, 35, 45, 55, 100]
labels = ['Young', 'Adult', 'Middle Age', 'Older Adult', 'Senior']
```

```python
data_read['Age'] = pd.to_numeric(data_read['Age'], errors='coerce')
data_read['Age_Group'] = pd.cut(data_read['Age'], bins=bins, labels=labels, right=False)
data_read
```

**Alligning the Columns**

```python
cols = list(data_read.columns)
cols.remove('Age_Group')
age_index = cols.index('Age')
cols.insert(age_index + 1, 'Age_Group')
data_read = data_read[cols]
data_read
```
